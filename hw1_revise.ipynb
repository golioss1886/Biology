```python
import matplotlib.pyplot as plt
```

```python
from Bio import SeqIO

ETC_all = open("etc0.fasta")
TC_all = open("transport.fasta")
ETC_seq = []
TC_seq = []
for i in SeqIO.parse(ETC_all, "fasta"):
    ETC_seq.append(i.seq)
ETC_all.close()
for i in SeqIO.parse(TC_all, "fasta"):
    TC_seq.append(i.seq)
TC_all.close()
print(len(ETC_seq))
print(len(TC_seq))
```

```python
x_axis = ['A','R','N','D','C','Q','E','G','H','I','L','K','M','F','P','S','T','W','Y','V']
aminos_cnt = {
    'A':0,'R':0,'N':0,'D':0,'C':0,'Q':0,'E':0,'G':0,'H':0,'I':0,'L':0,'K':0,'M':0,'F':0,'P':0,'S':0,'T':0,'W':0,'Y':0,'V':0
}
# print(type(ETC_seq))
for i in ETC_seq:
    for a in x_axis:
        aminos_cnt[a] += i.count(a)
print(aminos_cnt)
total_cnt = 0
for k, v in aminos_cnt.items():
    total_cnt += v
print(total_cnt)
y_axis = []
for i in list(aminos_cnt.values()):
    y_axis.append(i/total_cnt * 100)
print(y_axis)
```

```python
# For Complexs
from Bio import SeqIO
def get_list(file_path):
    bio_object = open(file_path)
    fasta_seq = []
    for i in SeqIO.parse(bio_object, "fasta"):
        fasta_seq.append(i.seq)
    bio_object.close()
    return fasta_seq

#C1 = open("etc1.fasta")
#C2 = open("etc2.fasta")
#C3 = open("etc3.fasta")
#C4 = open("etc4.fasta")
C1_seq = get_list("etc1.fasta")
C2_seq = get_list("etc2.fasta")
C3_seq = get_list("etc3.fasta")
C4_seq = get_list("etc4.fasta")
print(len(C1_seq))
print(len(C2_seq))
print(len(C3_seq))
print(len(C4_seq))
```

```python
def get_y_axis(myseq):
    x_axis = ['A','R','N','D','C','Q','E','G','H','I','L','K','M','F','P','S','T','W','Y','V']
    aminos_cnt = {
    'A':0,'R':0,'N':0,'D':0,'C':0,'Q':0,'E':0,'G':0,'H':0,'I':0,'L':0,'K':0,'M':0,'F':0,'P':0,'S':0,'T':0,'W':0,'Y':0,'V':0
}
    for i in myseq:
        for a in x_axis:
            aminos_cnt[a] += i.count(a)
    print(aminos_cnt)
    total_cnt = 0
    for k, v in aminos_cnt.items():
        total_cnt += v
    print(total_cnt)
    y_axis = []
    for i in list(aminos_cnt.values()):
        y_axis.append(i/total_cnt * 100)
    return y_axis

plt.clf()
C1_y = get_y_axis(C1_seq)
C2_y = get_y_axis(C2_seq)
C3_y = get_y_axis(C3_seq)
C4_y = get_y_axis(C4_seq)
```

```python
plt.plot(x_axis, C1_y, label = "Complex I")
plt.plot(x_axis, C2_y, label = "Complex II")
plt.plot(x_axis, C3_y, label = "Complex III")
plt.plot(x_axis, C4_y, label = "Complex IV")
plt.legend()
plt.xlabel('Amino acid residues') 
plt.ylabel('Composition (%)') 
plt.savefig("Complex.png")

```
